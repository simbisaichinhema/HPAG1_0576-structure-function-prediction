# 04. Active Site and Binding Pocket Detection (CASTp)

## 1. Objective
To identify potential active sites, binding pockets, and surface cavities within the predicted 3D structure of HPAG1_0576. Locating these regions is essential for functional annotation and identifying sites for potential molecular interactions or drug binding.

## 2. Methodology
The AlphaFold-predicted model (Rank 1, relaxed) was analyzed using **CASTp 3.0** (Computed Atlas of Surface Topography of proteins). CASTp uses a weighted Delaunay triangulation and Alpha complex method to identify and measure surface accessible pockets and interior voids. A probe radius of 1.4 Å was used to define the molecular surface.

## 4. Detailed Interpretation
The largest identified cavity, **Pocket 1**, is a substantial surface pocket with an area of **124.79 Å²** and a volume of **100.08 Å³**. This pocket is lined by a mix of hydrophobic (L90, I101, I168) and polar/charged residues (R100, K172, E175), suggesting a versatile binding environment. Interestingly, these residues are located within the conserved **Tipalpha domain** (identified in Steps 5 & 6), reinforcing the hypothesis that this region is the primary functional site of the protein. The presence of charged residues within the pocket suggests potential for electrostatic interactions with host cell receptors or other signaling molecules. **Pocket 2** also provides a significant surface area (51.47 Å²) and is located near the N-terminus, potentially serving as a secondary site for protein-protein interactions.

## 5. Conclusion
HPAG1_0576 contains well-defined surface pockets, with the largest cavity overlapping perfectly with the conserved functional domain. This structural organization is highly suggestive of an active role in molecular binding or signaling, supporting the protein's predicted role as a virulence factor in *H. pylori*.
