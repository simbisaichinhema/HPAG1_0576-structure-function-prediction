# 03. Physicochemical Characterization of HPAG1_0576

## 1. Objective
Physicochemical analysis is performed to define the fundamental biochemical identity of HPAG1_0576, providing essential insights into its stability, solubility, and charge distribution. Characterizing these properties is a prerequisite for understanding how the protein maintains its fold and function within the specific physiological environment of *Helicobacter pylori*.

## 2. Methodology
The primary physicochemical parameters were computed using the **ExPASy ProtParam** tool (via Biopython). This methodology derives quantitative data directly from the primary amino acid sequence, calculating properties such as molecular weight, isoelectric point, and hydropathicity based on established amino acid scales.

```python
import sys
import os
import pandas as pd
sys.path.append('../')
from src.utils import parse_fasta, calculate_physicochemical_properties, get_amino_acid_composition

# Load sequence
fasta_path = '../data/raw/HPAG1_0576.fasta'
record = parse_fasta(fasta_path)
sequence = record.seq

# Calculate properties
props = calculate_physicochemical_properties(sequence)
df_props = pd.DataFrame([props])
df_props.to_csv('../results/tables/physicochemical_properties.csv', index=False)
df_props
```

## 3. Results

| Parameter | Value | Biological Interpretation |
| :--- | :--- | :--- |
| Number of amino acids | 192 | Suggests a medium-sized, likely single-domain globular protein. |
| Molecular weight | 21.85 kDa | Typical for bacterial regulatory or accessory proteins. |
| Theoretical pI | 8.36 | Indicates a basic nature, carrying a net positive charge at physiological pH. |
| Instability index | 49.85 | Classifies the protein as unstable under *in vitro* experimental conditions. |
| Aliphatic index | 95.94 | High value suggests structural compactness and thermal stability. |
| GRAVY | -0.220 | Hydrophilic character consistent with a soluble, non-membrane protein. |

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Amino Acid Composition Visualization
comp = get_amino_acid_composition(sequence)
plt.figure(figsize=(12, 6))
sorted_comp = dict(sorted(comp.items()))
sns.barplot(x=list(sorted_comp.keys()), y=list(sorted_comp.values()), palette='viridis')
plt.title("Amino Acid Composition of HPAG1_0576", fontsize=14)
plt.ylabel("Percentage (%)")
plt.savefig("../results/figures/aa_composition.png", dpi=300)
plt.show()
```

## 4. Integrated Analysis
With a sequence length of 192 residues and a molecular weight of 21.85 kDa, HPAG1_0576 is likely a single-domain protein, a common architecture for bacterial signaling or binding factors. The theoretical pI of 8.36 indicates that the protein is basic, which may represent an adaptation for interaction with negatively charged molecular targets or a requirement for stability in the specific microenvironments of the *H. pylori* cell. Although the instability index of 49.85 predicts *in vitro* instability, this is potentially compensated by the high aliphatic index of 95.94, which points toward a dense hydrophobic core capable of providing significant thermal and structural resilience *in vivo*. The slightly positive charge balance—resulting from 24 positively charged versus 22 negatively charged residues—suggests an electrostatic surface capable of specific interactions. Furthermore, the GRAVY score of -0.220 confirms an overall hydrophilic nature, strongly suggesting that the protein is soluble and likely resides in the cytoplasm or periplasmic space.

## 5. Functional Inference
HPAG1_0576 appears to be a soluble, globular protein characterized by a stable aliphatic core and a hydrophilic, basic surface. This profile is indicative of a protein involved in molecular recognition or regulatory processes rather than a structural membrane component. The balanced charge distribution and predicted solubility suggest it may function as a cytoplasmic factor or a transiently associated binding partner within the *H. pylori* proteome.

## 6. Conclusion
The biochemical identity of HPAG1_0576 is that of a small, basic, and soluble protein with high predicted structural compactness. Its physicochemical profile confirms its suitability for downstream computational analyses, including domain identification and interaction mapping.
