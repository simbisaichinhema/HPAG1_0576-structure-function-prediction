# 01. Sequence Retrieval

## Objective
The objective of this notebook is to retrieve the amino acid sequence of **HPAG1_0576** from public databases (UniProt) using a search query and prepare it for downstream analysis.

## Input Data
- Search Query: `HPAG1_0576 Helicobacter pylori`

## Methodology
We use the `requests` library to fetch the FASTA sequence directly from UniProt's REST API search endpoint. The sequence is then parsed using `Biopython` and saved to the `data/raw/` directory.

```python
import requests
import os

# UniProt FASTA fetch using search query
url = "https://rest.uniprot.org/uniprotkb/search?query=HPAG1_0576+Helicobacter+pylori&format=fasta"

print(f"Fetching sequence from: {url}")
response = requests.get(url)

if response.status_code == 200:
    fasta_data = response.text
    
    # Ensure directory exists
    os.makedirs("../data/raw", exist_ok=True)
    
    # Save to file
    with open("../data/raw/HPAG1_0576.fasta", "w") as f:
        f.write(fasta_data)
    
    print("FASTA saved successfully to ../data/raw/HPAG1_0576.fasta")
else:
    print(f"Error fetching sequence: {response.status_code}")
```

```python
from Bio import SeqIO

# Parse and Display Metadata
try:
    records = list(SeqIO.parse("../data/raw/HPAG1_0576.fasta", "fasta"))
    if records:
        record = records[0]
        print(f"ID: {record.id}")
        print(f"Description: {record.description}")
        print(f"Sequence Length: {len(record.seq)} AA")
        print(f"Sequence Preview: {record.seq[:50]}...")
    else:
        print("No records found in FASTA file.")
except Exception as e:
    print(f"Error parsing FASTA: {e}")
```

## Results: Database Comparison

Below is a comparison of metadata retrieved from different sources.
